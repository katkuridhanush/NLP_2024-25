```python
import nltk
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize
def disambiguate_words(sentence):
  words=word_tokenize(sentence)
  disambiguated_words=[]
  for word in words:
    synset=lesk(sentence,word)
    if synset:
      disambiguated_words.append(synset.name().split('.')[0])
    else:
      disambiguated_words.append(word)
  return ' '.join(disambiguated_words)
  sentence="Hello I am Dhanush and this is my first assignment in NLP"
  disambiguated_sentence=disambiguate_words(sentence)
  print(disambiguated_sentence)
```
